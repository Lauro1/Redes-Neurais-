```python
import requests
from bs4 import BeautifulSoup
import os

def download_images(search_term, save_dir):
  search_url = f"https://www.google.com/search?q={search_term}&tbm=isch"
  response = requests.get(search_url)
  soup = BeautifulSoup(response.content, 'html.parser')
  images = soup.find_all('img', {'class': 'rg_i Q4LuWd'})
  count = 0
  for image in images:
      img_url = image['src']
      img_data = requests.get(img_url).content
      with open(os.path.join(save_dir, 'image_' + str(count) + '.jpg'), 'wb') as handler:
          handler.write(img_data)
      count += 1

# Exemplo de uso
download_images('coleira de cachorro pitbull', '/content/drive/MyDrive/imagens_produtos/coleira')
```

```python
import os
import cv2
from google.colab import drive
from google.colab.patches import cv2_imshow

# Montar o Google Drive
drive.mount('/content/drive')

# Definir o caminho para a pasta com as imagens
pasta_coleiras = '/content/drive/MyDrive/TreinamentoMachineLearning/coleira'

# Listar todos os arquivos na pasta
lista_arquivos = os.listdir(pasta_coleiras)

# Iterar sobre cada arquivo e carregar a imagem
for arquivo in lista_arquivos:
  # Construir o caminho completo para o arquivo
  caminho_completo = os.path.join(pasta_coleiras, arquivo)

  # Carregar a imagem
  try:
    imagem = cv2.imread(caminho_completo)
    print(f"Imagem carregada com sucesso: {caminho_completo}")

    # Exibir a imagem usando cv2_imshow
    cv2_imshow(imagem)
    cv2.waitKey(0)

  except cv2.error as e:
    print(f"Erro ao carregar a imagem {caminho_completo}: {e}")
```

```python
import requests
from bs4 import BeautifulSoup
import urllib.parse

termo_pesquisa = "cachorros com coleiras"
url = f"https://www.google.com/search?q={urllib.parse.quote(termo_pesquisa)}&tbm=isch"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

imagens = soup.find_all('img', class_='rg_i Q4LuWd')  # Adaptar a classe conforme a estrutura do HTML

for i, imagem in enumerate(imagens[:3]):
    img_url = imagem['src']
    img_data = requests.get(img_url).content
    with open(f'imagem_{i}.jpg', 'wb') as handler:
        handler.write(img_data)
```
